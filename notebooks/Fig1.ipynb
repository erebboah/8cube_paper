## Long function
Modified from decoupler https://github.com/saezlab/decoupler-py/blob/3dde7ee3804ef5d5d3bb88a97c212bb5aace4ffb/decoupler/plotting.py#L1535


```python
def my_plot_associations(data, 
                         uns_key, 
                         associations=None, 
                         cols=None, 
                         obs_annotation_cols=None, 
                         obsm_key=None, use_X=False,
                         layer=None, stat_col='p_adj', 
                         titles=['Scores', 'Stats'], 
                         sign_thr=0.01, n_factors=10,
                         cmap_stats='Purples', cmap_scores='BrBG', 
                         figsize=(5, 5), dpi=100):
    
    
    stats = data.uns[uns_key]
    stats.replace(0, np.finfo(float).eps, inplace=True)
    
    column_name = "PC"
    columns = ['{0}{1}'.format(column_name, 1 + x) for x in range(data.obsm[obsm_key].shape[1])]
    acts = pd.DataFrame(data.obsm[obsm_key],index=data.obs.index,columns=columns)
    
    tissue_order = ['CortexHippocampus',
                     'DiencephalonPituitary',
                     'Heart',
                     'GonadsFemale',
                     'Adrenal',
                     'Gastrocnemius',
                     'Kidney',
                     'Liver',
                     'GonadsMale',
                     ]
    genotype_order = ["AJ", "B6J", "129S1J", "NODJ", "NZOJ", "CASTJ", "PWKJ", "WSBJ"]

    tissue_order_map = {tissue: i for i, tissue in enumerate(tissue_order)}
    genotype_order_map = {genotype: i for i, genotype in enumerate(genotype_order)}

    acts['Tissue'] = data.obs['Tissue']  # Ensure Tissue column matches index
    acts['Genotype'] = data.obs['Genotype']
    acts['Sex'] = data.obs['Sex']  # Ensure Sex column is available
    acts['Tissue_Order'] = acts['Tissue'].map(tissue_order_map)
    acts['Genotype_Order'] = acts['Genotype'].map(genotype_order_map)

    def assign_sex_order(row):
        if row['Tissue'] in ['GonadsMale', 'GonadsFemale']:
            return 0 if row['Sex'] == 'Female' else 1  # No alternation for gonads
        else:
            return 0 if row['Sex'] == 'Female' else 1  # Alternate Male/Female elsewhere

    acts['Sex_Order'] = acts.apply(assign_sex_order, axis=1)

    acts_sorted_full = acts.sort_values(by=['Tissue_Order', 'Genotype_Order', 'Sex_Order'])
    acts_sorted = acts_sorted_full.drop(columns=['Tissue_Order', 'Genotype_Order', 'Sex_Order',
                                           'Tissue', 'Genotype', 'Sex'])

    acts = acts_sorted

    col_order = stats['factor'].unique()
    stats = stats.pivot(index='factor', columns='variable', values=stat_col).T.loc[:, col_order]
    stats.index.name = None
    stats.columns.name = None
    stats = -np.log10(stats)

    stats = stats.loc[:, stats.columns[:n_factors]].copy()
    #print(stats)
    
    acts = acts.loc[:, stats.columns].copy()

    h1 = ma.Heatmap(stats, cmap=cmap_stats, width=4, height=1, name="h1")
    h1.add_title(top=titles[1], align="center")
    h1.add_legends()
    #h1.add_dendrogram("left")
    h1.add_right(mp.Labels(stats.index, ))
    sign_msk = stats.values > -np.log10(sign_thr)
    layer = mp.MarkerMesh(sign_msk, marker='*', label="{0} < {1}".format(stat_col, str(sign_thr)))
    h1.add_layer(layer, name='sign')

    acts_sorted_full['Tissue_colors'] = acts_sorted_full['Tissue'].map(tissue_dict)
    acts_sorted_full['Genotype_colors'] = acts_sorted_full['Genotype'].map(geno_dict)
    acts_sorted_full['Sex_colors'] = acts_sorted_full['Sex'].map(sex_dict)

    h2 = ma.Heatmap(acts.values, cmap=cmap_scores, width=0.4, height=4, name="h2")
    h2.add_title(top=titles[0], align="center")
    h2.add_legends()

    variance_ratio = my_pdata.uns["pca"]["variance_ratio"]
    variance_labels = [f"PC{i + 1} ({round(v * 100, 2)}%)" for i, v in enumerate(variance_ratio)]
    h2.add_bottom(mp.Labels(variance_labels[:acts.shape[1]], rotation=90, size=12))  # Ensure correct length

    for cat_col, color_dict in zip(['Tissue', 'Genotype', 'Sex'], [tissue_dict, geno_dict, sex_dict]):
        color_col = cat_col + '_colors'
        cats = acts_sorted_full[cat_col].values  # Original categories
        cat_colors = acts_sorted_full[color_col].values  # Corresponding colors
        cat_colors = mp.Colors(cats, palette=color_dict, label=cat_col)  # Ensure correct legend order
        h2.add_right(cat_colors, pad=0.1, size=0.2)


    c = (h1 / .2 / h2)
    c.add_legends(side="right",
                  stack_by='row', stack_size=2)
    c.render()

    hax = c.get_ax(board_name='h1', ax_name='h1')
    border = Rectangle((0, 0), 1, 1, fill=False, ec=".1", lw=2, transform=hax.transAxes)
    hax.add_artist(border)

    hax = c.get_ax(board_name='h2', ax_name='h2')
    border = Rectangle((0, 0), 1, 1, fill=False, ec=".1", lw=2, transform=hax.transAxes)
    hax.add_artist(border)

    fig = c.figure
    fig.set_figwidth(figsize[0])
    fig.set_figheight(figsize[1])
    fig.set_dpi(300)
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

def create_standalone_legend(color_dict, label, order, location='upper left'):
    ordered_dict = {key: color_dict[key] for key in order if key in color_dict}
    legend_handles = [mpatches.Patch(color=color) for color in ordered_dict.values()]
    legend_labels = list(ordered_dict.keys())  # Legend items in the correct order

    fig, ax = plt.subplots(figsize=(3, 1))
    legend = ax.legend(legend_handles, legend_labels, title=label, loc=location, fontsize=10, title_fontsize=12)
    legend.set_title(label, prop={'weight': 'bold'})  # Make the legend title bold
    ax.axis('off')  
    fig.set_dpi(300)
    plt.savefig(f"../fig1/{label}_legend.png",bbox_inches='tight', dpi = 300)
    plt.show()

tissue_order = ['CortexHippocampus',
                     'DiencephalonPituitary',
                     'Heart',
                     'GonadsFemale',
                     'Adrenal',
                     'Gastrocnemius',
                     'Kidney',
                     'Liver',
                     'GonadsMale',
                     ]
genotype_order = ["AJ", "B6J", "129S1J", "NODJ", "NZOJ", "CASTJ", "PWKJ", "WSBJ"]

create_standalone_legend(tissue_dict, 'Tissue', tissue_order, location='upper left')
create_standalone_legend(geno_dict, 'Genotype', genotype_order, location='upper left')
create_standalone_legend(sex_dict, 'Sex', list(sex_dict.keys()), location='upper left')  # Sex doesn't need specific order

strain_dict = {'Lab': 'darkgrey', 'Wild': '#d3a021'}
create_standalone_legend(strain_dict, 'Strain type', list(strain_dict.keys()), location='upper left')  # Sex doesn't need specific order
```

```python
import scanpy as sc
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
import matplotlib.pylab as plt
import PyComplexHeatmap as pch
```

## Read in (old) clustering results

```python
all_tissues_data = []

for tissue in tissues:
    # Read and filter the data
    chisq_results = pd.read_csv(
        f"/share/crsp/lab/seyedam/share/igvf_pipeline/figures/subclustering_subtypes/{tissue}/{tissue}_proportion_results_fixed_prop{prop}_res{res}_fixed_hvgs_1217.csv", 
        index_col=0
    )
    chisq_results = chisq_results[chisq_results['subtype'] != 'low quality']

    # Group and calculate percentages
    grouped_df = chisq_results.groupby(['Genotype', 'tissue']).agg({
        'ncells_specific_cluster': 'sum',
        'ncells_total': 'sum'
    }).reset_index()
    grouped_df['percentage_cells'] = (grouped_df['ncells_specific_cluster'] / grouped_df['ncells_total']) * 100

    # Append to the list
    all_tissues_data.append(grouped_df)

```

```python
merged_df = pd.concat(all_tissues_data)

plot_df = merged_df.pivot_table(
    index=['tissue'], 
    columns='Genotype',         
    values='percentage_cells',   
    fill_value=0)

```

```python
plt.figure(figsize=(3.1, 3))

genos = merged_df['Genotype'].unique().tolist()
df_col = pd.DataFrame({
    "Genotype": genos, 
    "Strain type": ["Wild" if x in ["CASTJ","PWKJ","WSBJ"] else "Lab" for x in genos]
})

genotype_categories = ['AJ','B6J','129S1J','NODJ','NZOJ','CASTJ','PWKJ','WSBJ']
df_col['Genotype'] = pd.Categorical(df_col['Genotype'], categories=genotype_categories, ordered=True)
df_col.sort_values(['Genotype'], inplace=True)
df_col = df_col.set_index("Genotype", drop=False)

col_ha = pch.HeatmapAnnotation(Genotype=pch.anno_simple(df_col['Genotype'],colors=geno_dict),
                               Type=pch.anno_simple(df_col['Strain type'],colors=strain_dict),
                               label_kws = {'visible': False},
                               axis=1)

cm = pch.ClusterMapPlotter(data=plot_df.loc[df_row.index.tolist(),df_col.index.tolist()],
                           show_rownames = True, 
                           show_colnames = True, 
                           cmap = "mako",
                           label = "% nuclei clustered\nby genotype",
                           top_annotation=col_ha, 
                           left_annotation=left_ha,
                           row_cluster=False,
                           col_cluster=False,
                           vmin = 0,
                           vmax = 100,
                           yticklabels_kws={'labelsize': 14},
                           xticklabels_kws={'labelsize': 14} 
                          )

plt.savefig(f"../fig1/subclustering_tissue_level_hmap.png", 
            dpi=300, bbox_inches="tight", format="png")


plt.show()
```
